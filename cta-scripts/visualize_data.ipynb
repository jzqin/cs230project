```python
import matplotlib
import matplotlib.pyplot as plt
import imageio as imgio
import numpy as np
import pandas as pd
import nrrd
```

```python
# maskName = '/data2/yeom/ky_aneur/segmentation/segmentations_181128/xxxx.nrrd'
# predName = '/data2/yeom/ky_aneur/results/dev/train2-resnet50-ppm_deepsup-ngpus2-batchSize4-imgMaxSize1000-paddingConst8-segmDownsampleRate8-LR_encoder0.02-LR_decoder0.02-epoch5/IM-0028-0{}.png'
maskName = '/data2/yeom/ky_aneur/segmentation/segmentations_181128/xxxx.nrrd'
predName = '/data2/yeom/ky_aneur/results/dev/train3-resnet50-ppm_deepsup-ngpus2-batchSize4-imgMaxSize1000-paddingConst8-segmDownsampleRate8-LR_encoder0.02-LR_decoder0.02-epoch5/IM-0008-0{0:0=3d}.png'


mask = nrrd.read(maskName)
mask = mask[0].astype(np.uint8)

maskSums = np.sum(mask, axis=0)
maskSums = np.sum(maskSums, axis=0)
maskPos = np.where(maskSums > 0)
maskPos = np.squeeze(maskPos)
print(maskPos)

for i in maskPos:
    img = imgio.imread(predName.format(i))
    plt.imshow(img[:, :512, :], 'gray', interpolation='none')
    plt.show()
    
    plt.imshow(img[:, :512, :], 'gray', interpolation='none')
    plt.imshow(img[:, -1024:-512, :] - 120, interpolation='none', alpha=0.55)
    plt.show()
    
    plt.imshow(img[:, :512, :], 'gray', interpolation='none')
    plt.imshow(img[:, -512:, :] - 120, interpolation='none', alpha=0.55)
    plt.show()


    print(np.sum(img[:, -1024:-512, :] - 120))
    print(np.sum(img[:, -512:, :] - 120))
```
