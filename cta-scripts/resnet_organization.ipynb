```python
import json
import pydicom as dicom
import numpy as np
import os
import matplotlib.pyplot as plt
import imageio as imgio
```

```python
filePath = '/data2/yeom/ky_aneur/sah_old/Aguirre Ana Zuleima/6073058/H-N ANGIO 1.25MM/'
output = '/home/ky_aneur/cs230/resnet/test/'
for f in os.listdir(filePath):
    imgPath = filePath + f
    with open(imgPath, 'rb') as d:
        dcmData = dicom.dcmread(d)
    # import pdb; pdb.set_trace()
    dcmImg = dcmData.pixel_array
    fileInfo = {}
    fileInfo['dbInfo'] = {'frameID': -1, 'vID': 'CS230'}
    fileInfo['width'] = dcmImg.shape[1]
    fileInfo['fpath_img'] = '/data2/yeom/ky_aneur/test/png/' + f[:-4] + '.png'
    fileInfo['ade_scene'] = 'brain'
    fileInfo['height'] = dcmImg.shape[0]
    fileInfo['dbName'] = 'CS230'
    fileInfo['fpath_segm'] = '/data2/yeom/ky_aneur/test/png/' + f[:-4] + '_mask.png'
    with open(output + 'test.odgt', 'a') as results:
        json.dump(fileInfo, results)
        results.write('\n')
```

```python
# convert from dcm to png
filePath = '/data2/yeom/ky_aneur/sah_old/Aguirre Ana Zuleima/6073058/H-N ANGIO 1.25MM/'
output = '/data2/yeom/ky_aneur/test/png/'
for f in os.listdir(filePath):
    imgPath = filePath + f
    with open(imgPath, 'rb') as d:
        dcmData = dicom.dcmread(d)
    # import pdb; pdb.set_trace()
    dcmImg = dcmData.pixel_array
    dcmCopy = np.zeros((dcmImg.shape[0], dcmImg.shape[1]))
    plt.imsave(output + f[:-3] + 'png', dcmImg)
    imgio.imsave(output + f[:-4] + '_mask.png', dcmCopy)
```
